```python
%matplotlib inline
import dlib
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# 1

#  
MODEL_PATH = "./data/models/shape_predictor_68_face_landmarks.dat"

#  
imageFilename = 'data/images/laughing-couple.jpg'

# 
img= cv2.imread(imageFilename)
print("imgshape", img.shape)
plt.imshow(img[:,:,::-1])
plt.show()

# 
faceDetector = dlib.get_frontal_face_detector()
landmarkDetector = dlib.shape_predictor(MODEL_PATH)
faceRects = faceDetector(img, 1)
print("Number of faces detected: ",len(faceRects))

img1 =  img.copy()
drawBlur_copy = img.copy()

for i in range(0, len(faceRects)):
    #   
    x1 = faceRects[i].left()
    y1 = faceRects[i].top()
    x2 = faceRects[i].right()
    y2 = faceRects[i].bottom()
    
    
    # 
    face_cut = drawBlur_copy[y1:y2+1, x1:x2+1]
    plt.figure(figsize=(7,3))
    plt.subplot(121); plt.imshow(face_cut[:,:,::-1])
    # 
    drawblur = cv2.GaussianBlur(face_cut, (0, 0), 10)
    plt.subplot(122); plt.imshow(drawblur[:,:,::-1])
    
    #  
    drawBlur_copy[y1:y2+1, x1:x2+1] = drawblur
    cv2.rectangle(img1, (x1, y1), (x2, y2), (0, 0, 255), 2)
    
    
    
plt.figure(figsize=(8,10))
plt.subplot(211);
plt.imshow(img1[:,:,::-1])
plt.title('faces Rects')
plt.subplot(212);
plt.imshow(drawBlur_copy[:,:,::-1])
plt.title('face blur')
```

```python
### 2
# 
img2= cv2.imread(imageFilename)
plt.imshow(img2[:,:,::-1])
plt.show()

# 0.5 
threshold = 0.5
for i in range(0, len(faceRects)):
  # 
  landmark = landmarkDetector(img2, faceRects[i])
  print('(landmark.part(54).x, landmark.part(54).y): '
        ,(landmark.part(54).x , landmark.part(54).y ))
  #  
  chicLength = np.sqrt(np.power((landmark.part(8).x - landmark.part(57).x),2)
                       +np.power((landmark.part(8).y - landmark.part(57).y),2))
  # 
  mouthLength = np.sqrt(np.power((landmark.part(51).x - landmark.part(57).x),2) 
                        +np.power((landmark.part(51).y - landmark.part(57).y),2))
  #   /        
  if(mouthLength / chicLength >= threshold):
    #54   
    cv2.putText(img2, "Smile", (landmark.part(54).x, landmark.part(54).y), 
                0, 0.8, (0, 0, 255), 2, cv2.LINE_8)

plt.imshow(img2[:,:,::-1]);
plt.title("Smile")
```

```python
#  
frame = img.copy()

# 
threshold = 0.5

#  
heart_img = cv2.imread('./data/images/heart-icon.png') 
print('heart-icon.png', heart_img .shape);

# 
gray_mask = cv2.cvtColor(heart_img, cv2.COLOR_BGR2GRAY) 

#  255 
ret,h_mask = cv2.threshold(gray_mask, 110, 255, cv2.THRESH_BINARY_INV)

#3 
heart_mask = cv2.merge((h_mask, h_mask, h_mask))

# 
mask = cv2.bitwise_and(heart_img, heart_mask)

# ,  
heartHeight, heartWidth = heart_img.shape[:2] 


for i in range(0, len(faceRects)):
  # LANDMARK 
  landmark = landmarkDetector(frame, faceRects[i])
  chicLength = np.sqrt(np.power((landmark.part(8).x - landmark.part(57).x),2) +np.power((landmark.part(8).y - landmark.part(57).y),2))
  mouthLength = np.sqrt(np.power((landmark.part(51).x - landmark.part(57).x),2) +np.power((landmark.part(51).y - landmark.part(57).y),2))
  #SMILE 
  if(mouthLength / chicLength >= threshold):
        
    # 
    heart_x = landmark.part(54).x
    heart_y = landmark.part(54).y
    
    #  =     
    resized_heartHeight = landmark.part(57).y - landmark.part(51).y
    
    # =      .
    resized_heartWidth = np.int(heartWidth*(resized_heartHeight/heartHeight))
    
    
    #  
    frame_roi = frame[heart_y:heart_y+resized_heartHeight, heart_x:heart_x+resized_heartWidth] 
    
    #frame     
    frameMask = cv2.bitwise_not(heart_mask)
    
    #     
    heart_Resize = cv2.resize(heart_img, (resized_heartWidth, resized_heartHeight))
    heartMask_Resize = cv2.resize(heart_mask, (resized_heartWidth, resized_heartHeight))
    frameMask_Resize = cv2.resize(frameMask, (resized_heartWidth, resized_heartHeight))

    #   
    frameMask_inv = cv2.bitwise_and(frame_roi, frameMask_Resize)
    heart_Resize =  cv2.bitwise_and(heart_Resize, heartMask_Resize)
    #  
    print("frame[heart_y:heart_y+resized_heartHeight, heart_x:heart_x+resized_heartWidth] .shape", frame[heart_y:heart_y+resized_heartHeight, heart_x:heart_x+resized_heartWidth].shape)
    print("heart_Resize.shape", heart_Resize.shape)
    print("frameMask_inv.shape", frameMask_inv.shape)
    #frame[heart_y:heart_y+resized_heartHeight, heart_x:heart_x+resized_heartWidth] =cv2.addWeighted(heart_Resize, 1, frameMask_inv, 1, 0)
    frame[heart_y:heart_y+resized_heartHeight, heart_x:heart_x+resized_heartWidth] = cv2.add(heart_Resize, frameMask_inv)
    
    
    # 
    print(heartMask_Resize.shape)
    plt.figure(figsize=(6,6))
    plt.subplot(341); plt.imshow(h_mask[:,:]); plt.title('h_mask')
    plt.subplot(342); plt.imshow(heart_mask[:,:,::-1]); plt.title('heart_mask')
    plt.subplot(343); plt.imshow(heart_Resize[:,:,::-1]); plt.title('resized heart')
    plt.subplot(344); plt.imshow(frameMask[:,:,::-1]); plt.title('frameMask')
    plt.subplot(345); plt.imshow(frameMask_Resize[:,:,::-1]); plt.title('frameMask_Resize')
    plt.subplot(346); plt.imshow(frameMask_inv[:,:,::-1]); plt.title('frameMask_inv')
    plt.subplot(347); plt.imshow(frame_roi[:,:,::-1]); plt.title('frame_roi')
    
plt.figure(figsize=(6,6))
plt.imshow(frame[:,:,::-1]); plt.title('frame_masked')
    

```

```python
# project #4 - 2 :   
import math

frame2 = img.copy()
h, w = frame2.shape[:2]

# sunglasses    
sunglass_img = cv2.imread('data/images/sunglasses.png')  # sunglasses_image

# , 
sunglass_h, sunglass_w = sunglass_img[:2]
print('sunglass_img :', sunglass_img.shape, sunglass_img.dtype)
print('   min, max :', np.amin(sunglass_img), np.amax(sunglass_img))

# 
plt.figure(figsize=(12,12))
plt.subplot(121); plt.imshow(sunglass_img); plt.title('sunglass_img')

scaling_factor = 0.5 

#  
#frame2 = cv2.resize(frame2, None, fx = scaling_factor,fy=scaling_factor, interpolation=cv2.INTER_AREA)
print('frame2(resized) :', frame2.shape, frame2.dtype)
vh, vw = frame2.shape[:2]

#  
gray = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY) 

for i in range(0, len(faceRects)):
    
    #  
    landmark = landmarkDetector(frame2, faceRects[i])
    
    #   
    x = faceRects[i].left()
    y = faceRects[i].top()
    w = faceRects[i].right()
    h = faceRects[i].bottom()
    print("x, y, h, w", x, y, h, w)
    # 
    right_eye_x = landmark.part(36).x # right eye x
    right_eye_y = landmark.part(36).y # right eye y
    left_eye_x = landmark.part(45).x # left eye x
    left_eye_y = landmark.part(45).y # left eye y
    eyes_point = [(right_eye_x, right_eye_y),(left_eye_x, left_eye_y)]
    
    Rect = frame2.copy()
    cv2.rectangle(Rect, (x, y), (w, h), (0, 0, 255), 1)
    cv2.circle(Rect, eyes_point[0], 1, (0,0,255), -1)
    cv2.circle(Rect, eyes_point[1], 1, (0,0,255), -1)
    plt.figure(figsize=(13,13))
    plt.subplot(121); plt.imshow(Rect[:,:,::-1]); plt.title('face'+np.str(i))
    
    #   
    eye_distance = abs(eyes_point[1][0] - eyes_point[0][0])
    
    #  
    faceCenter = (int((eyes_point[0][0] + eyes_point[1][0])/2),
                  int((eyes_point[0][1]+eyes_point[1][1])/2))
   
    #    2.12 
    sunglasses_width = 2.12 * eye_distance
    scaling_factor = sunglasses_width / w
    print('eye_distance={},  scaling_factor={:.3f}'.format(eye_distance, scaling_factor))
    
    #  
    overlay_sunglasses = cv2.resize(sunglass_img, None, fx=0.054, fy=0.062, interpolation=cv2.INTER_AREA)
    print('overlay_sunglasses(resized) :', overlay_sunglasses.shape, overlay_sunglasses.dtype)
    h, w = overlay_sunglasses.shape[:2]
    print("x, y, h, w resize", x, y, h, w)

    #  
    frame_roi = frame2[y:y+h, x:x+w]
    rh, rw = frame_roi.shape[:2]
    print('frame_roi :', frame_roi.shape, frame_roi.dtype)

    #  
    roi_gray = gray[y:h, x:w] 
    roi_color = frame2[y:h, x:w] 

    #   0 
    img_zero = np.zeros_like(frame_roi)
    print('img_zero :', img_zero.shape, img_zero.dtype)
    plt.figure(figsize=(20,20))
    plt.subplot(441); plt.imshow(img_zero[:,:]); plt.title('img_zero'+np.str(i))
    
    gray_overlay_sunglassess = cv2.cvtColor(overlay_sunglasses, cv2.COLOR_BGR2GRAY) 
    print('gray_overlay_sunglassess :', gray_overlay_sunglassess.shape, gray_overlay_sunglassess.dtype)
    


    # 5.     
    dX = abs(left_eye_x- right_eye_x)
    dY = abs(left_eye_y- right_eye_y)
    angle = np.degrees(np.arctan2(dY, dX))-7
    print('dx, dy, angle', dX, dY, angle)    
    
    glassesCenter = [x+int(w*0.5), y+int(h*0.5)]
    print('glassesCenter', glassesCenter)
    
   
    #     180  255  
    ret, mask = cv2.threshold(gray_overlay_sunglassess, 253, 255, cv2.THRESH_BINARY_INV) 
    

    # ,  
    sunglassHeight, sunglassWidth = mask.shape[:2]

    # ,   
    R_sunglass_x = np.int(sunglassWidth/8)
    L_sunglass_x = np.int(7*sunglassWidth/8)
    R_sunglass_y = np.int(4*sunglassHeight/9)  
    L_sunglass_y = np.int(4*sunglassHeight/9)  
    sunglass_point = [(R_sunglass_x, R_sunglass_y), (L_sunglass_x, L_sunglass_y)]
    
    
    #  , 
    lensLeft = [R_sunglass_x, R_sunglass_y]
    lensRight = [L_sunglass_x, L_sunglass_y]
    print('lensLeft, lensRight', lensLeft, lensRight)
    
    # x, y 
    distance_center_x = abs(faceCenter[0] - glassesCenter[0])
    distance_center_y = abs(faceCenter[1] - glassesCenter[1])
    print('distance_center_x',distance_center_x)
    print('distance_center_y',distance_center_y)
    x += distance_center_x
    y += distance_center_y
    
    # 
    matrix = cv2.getRotationMatrix2D((w/2, h/2),angle, 1)
    
    mask = cv2.warpAffine(mask, matrix, (w, h))
    overlay_sunglasses = cv2.warpAffine(overlay_sunglasses, matrix, (w, h))

    #   
    mask_inv = cv2.bitwise_not(mask) 
    print('mask_inv :', mask_inv.shape, mask_inv.dtype)

    mask_3ch= cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
    print(' mask_3ch', mask_3ch.shape);
    plt.subplot(442); plt.imshow(mask_3ch[:,:]); plt.title('mask_3ch'+np.str(i))
    
    mask_inv_3ch= cv2.cvtColor(mask_inv, cv2.COLOR_GRAY2BGR)
    print(' mask_inv_3ch', mask_inv_3ch.shape);
    plt.subplot(443); plt.imshow(mask_inv_3ch[:,:]); plt.title('mask_inv_3ch'+np.str(i))
    
    
    # 
    masked_glasses = cv2.bitwise_and(overlay_sunglasses, overlay_sunglasses, mask=mask)
    plt.subplot(444); plt.imshow(masked_glasses[:,:]); plt.title('masked_glasses'+np.str(i))

    
    #       
    # Use the inverse mask to get the remaining part of the image 
    masked_frame = cv2.bitwise_and(frame_roi, frame_roi, mask=mask_inv)
    print(' masked_frame', masked_frame.shape);
    plt.subplot(445); plt.imshow(masked_frame[:,:,::-1]); plt.title('masked_frame'+np.str(i))
 
    # 
    masked_frame_eye = cv2.bitwise_and(frame_roi, frame_roi, mask=mask) 
    print(' masked_frame_eye', masked_frame_eye.shape);
    plt.subplot(446); plt.imshow(masked_frame_eye[:,:,::-1]); plt.title('masked_frame_eye'+np.str(i))
 

    frame2[y:y+h, x:x+w] = cv2.addWeighted(masked_glasses, 0.3, masked_frame_eye, 0.7, 0)
    print('masked_frame_eye masked_face     ');
    plt.subplot(447); plt.imshow(frame2[:,:,::-1]); plt.title('Alpha Blending 1 - '+np.str(i))
    #     
    frame2[y:y+h, x:x+w] = cv2.addWeighted(masked_frame, 1, frame2[y:y+h, x:x+w], 1, 0)
    print(' masked_frame frame[y:y+h, x:x+w] ');
    plt.subplot(448); plt.imshow(frame2[:,:,::-1]); plt.title('Alpha Blending 2 -'+np.str(i))

plt.figure(figsize=(12,12))
plt.subplot(121); plt.imshow(frame2[:,:,::-1]); plt.title('OUTPUT')
```
